```python
def dcf_fade_period(self, date, analysis, fade_rate, lgr):
    explicit_period = self.dcf_explicit_period(date, analysis)
    initial_ric = explicit_period['RIC'].mean()
    initial_sales = explicit_period['Sales'][-1]
    initial_date = explicit_period.index[-1]
    initial_nopat = explicit_period['NOPAT']
    wacc = self.calc_wacc(date, analysis)
    initial_difference_ric = wacc - initial_ric
    initial_difference_growth = initial_sales_growth - lgr
    if initial_difference_growth > 0:
        direction_g = -1
    else:
        direction_g = 1
    # creating the series for the RIC fade 
    i = 0
    ric_list = []
    g_list = []
    date_list = []
    while round(initial_ric, 2) != round(wacc, 2):
        increment_ric = fade_rate * initial_difference_ric 
        initial_ric = initial_ric + increment_ric
        increment_g = fade_rate * initial_difference_growth
        initial_sales_growth = initial_sales_growth + increment_g * direction_g
        initial_difference_ric = wacc - initial_ric
        initial_difference_growth = initial_sales_growth - lgr
        initial_date = initial_date + datetime.timedelta(days=365)
        g_list.append(initial_sales_growth)
        ric_list.append(initial_ric)
        date_list.append(initial_date)
    ric_series = pd.Series(ric_list, index=date_list)
    growth_series = pd.Series(g_list, index=date_list)
    new_nopats = []
    new_deltanoas = []
    discount_rates = []
    for i in range(len(date_list)):
        date = date_list[i]
        new_nopat = initial_nopat + initial_deltanoa * estimated_RIC
        new_deltanoa = initial_nopat * growth.get(date)/ric.get(date)
        new_discount_rate = (1/(1 + wacc) ** (5 + i))
        new_nopats.append(new_nopat)
        new_deltanoas.append(new_deltanoa)
        initial_nopat = new_nopat
        initial_deltanoa = new_deltanoa
        
    
    
    


dcf_fade_period(cd.get('AAPL'), end_frm, analysis, .1, .019)
```

```python
# idea: update stock dataframe by each company individually
# build up the dataframe of new data by each company, this way can control for errors caused by specific tickers
# add this dataframe to existign one 
# challenges: would this impeded the speed of the programming? Use multiprocessing to speed up program
# list of tickers divided up into n lists 
# for each list, use pandas datareader to retrieve df for each company
# for each list, append the dataframes in a way to form a larger dataframe 
# once all the lists are done, append the dataframes to get the dataframe of update info
# add update info dataframe to existing dataframe

# source: https://www.geeksforgeeks.org/break-list-chunks-size-n-python/
def divide_chunks(l, n): 
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n]
        
        

def get_update_dataframe(start, company_dict):
    tickers = list(company_dict.keys())
    end = datetime.date.today()
    d = {}
    for ticker in tickers:
        df = pdr.DataReader([ticker], 'yahoo', start, end)
        d.update( df.to_dict() )
    return pd.DataFrame(d)

update_df = get_update_dataframe(datetime.date(day=13, month=9, year=2019), cd)
```
